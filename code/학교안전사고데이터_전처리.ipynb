데이터 병합 후 상관관계 분석
- 학급수_계
- 학생수_총계_계
- 학생수_총계_남
- 학생수_총계_여
- 학급당\n 학생수
- 교원1인당 학생수
- 학생\n1인당\n교지면적

(21~23년의 경우, '일반학급_학급수', '특수학급_학급수', '일반학급_학생수_계', '특수학급_학생수_계' 컬럼 추가)

(2019년)

```python
#grouby 전 확인
print(acci_df_19['지역'].unique())
print(acci_df_19['지역'].nunique())

print(school_df_19['시도'].unique())
print(school_df_19['시도'].nunique())
```

```python
#학교안전사고데이터 지역별 groupby 후 count
acci_19_loccnt = acci_df_19[['구분', '지역']].groupby('지역').count()
acci_19_loccnt
```

```python
#학교, 학과별 데이터셋 지역별 groupby 후 mean
school_19_gpmean = school_df_19[['시도', '학급수_계', '학생수_총계_계', '학생수_총계_남', '학생수_총계_여', '학급당\n 학생수', '교원1인당 학생수', '학생\n1인당\n교지면적']].groupby('시도').mean()
school_19_gpmean
```

```python
acci_corr_19 = pd.concat([acci_19_loccnt, school_19_gpmean], axis=1)
acci_corr_19
```

```python
#사고 발생 수와 컬럼 간 상관관계(19년)
acci_corr_19.corr()
#사고 발생 수와 모두 약한 상관관계를 보임
```
